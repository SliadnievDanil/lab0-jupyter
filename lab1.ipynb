```python
import numpy as np
import csv

def load_column_data(filename, column_name):
    """
    Функція для завантаження даних зі стовпця з CSV файлу.

    Parameters:
    filename (str): Шлях до CSV файлу.
    column_name (str): Назва стовпця, значення якого потрібно завантажити.

    Returns:
    np.ndarray: Масив numpy з даними стовпця.
    """
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        headers = next(reader)  # Пропускаємо заголовки
        column_index = headers.index(column_name)  # Знаходимо індекс стовпця
        data = np.array([int(row[column_index]) for row in reader if row[column_index] != ''])
    return data


filename = 'russia_losses_equipment.csv'
column_name = 'tank'  # Приклад для стовпця танків

# Виклик функції і отримання масиву numpy
tank_data = load_column_data(filename, column_name)

# Виводимо результат
print(tank_data)

```
