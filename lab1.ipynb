```python
import csv
import numpy as np

def get_column_data(filename, column_name):
    with open(filename, mode='r', encoding='utf-8') as file:
        reader = csv.reader(file)
        headers = next(reader)  
        column_index = headers.index(column_name)  
        column_data = [int(row[column_index]) for row in reader if row[column_index].isdigit()]
    return np.array(column_data)  

file_path = 'russia_losses_equipment.csv'  
column_name = 'aircraft'  
aircraft_losses = get_column_data(file_path, column_name)

print(aircraft_losses)
```

```python
file_path = 'russia_losses_equipment.csv'

aircraft_losses = get_column_data(file_path, "aircraft")

print("Aircraft loss array : {}:".format(len(aircraft_losses)))
print(aircraft_losses)
```

```python
import numpy as np
import pandas as pd

start_date = pd.to_datetime('2022-02-24')
daily_losses = np.diff(aircraft_losses)
dates = pd.date_range(start=start_date, periods=len(daily_losses))

print("Daily aircraft losses:")
for date, loss in zip(dates, daily_losses):
    print(f"{date.date()}: {loss} aircraft")
```

```python
largest_losses = np.sort(daily_losses)[:3]

print("Three highest daily aircraft losses:")
print(largest_losses)
```

```python
summer_losses = np.sum(np.diff(aircraft_losses[151:243]))

print("Number of aircraft shot down in the summer of 2022:", summer_losses)
```

```python
average_last_300_days = np.mean(daily_losses[-300:])

print("Average aircraft losses over the last 300 days:", average_last_300_days)
```

```python
import matplotlib.pyplot as plt

aircraft_losses_first_year = aircraft_losses[:365]

plt.figure(figsize=(8, 16), dpi=100)
plt.plot(aircraft_losses_first_year, linestyle='--', color='b', label='Aircraft losses')
plt.title('Aircraft losses during the first year of the war')
plt.xlabel('Day')
plt.ylabel('Number of aircraft')

plt.legend()
plt.grid(True, linestyle='-.', color='gray')
plt.savefig('aircraft_losses_year.png')

plt.show()

print("Values for the graph of aircraft losses for the first year of the war: {}:".format(len(aircraft_losses_first_year)))
print(aircraft_losses_first_year)
```
