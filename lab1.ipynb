```python
import csv
import numpy as np

def get_column_data(filename, column_name):
    with open(filename, mode='r', encoding='utf-8') as file:
        reader = csv.reader(file)
        headers = next(reader)  
        column_index = headers.index(column_name)  # Знаходимо індекс потрібного стовпця
        column_data = [float(row[column_index]) for row in reader if row[column_index].replace('.', '', 1).isdigit()]  # Збираємо дані стовпця
    return np.array(column_data)  # Перетворюємо список в numpy масив

file_path = 'russia_losses_equipment.csv'  
column_name = 'aircraft'  
aircraft_losses = get_column_data(file_path, column_name)

print(aircraft_losses)
```

```python
file_path = 'russia_losses_equipment.csv'

aircraft_losses = get_column_data(file_path, "aircraft")

print("Aircraft loss array : {}:".format(len(aircraft_losses)))
print(aircraft_losses)
```

```python
import numpy as np
import pandas as pd

start_date = pd.to_datetime('2022-02-24')
daily_losses = np.diff(aircraft_losses)
dates = pd.date_range(start=start_date, periods=len(daily_losses))

print("Daily aircraft losses:")
for date, loss in zip(dates, daily_losses):
    print(f"{date.date()}: {loss} aircraft")
```

```python
largest_losses = np.sort(daily_losses)[:3]

print("Three highest daily aircraft losses:")
print(largest_losses)
```

```python
summer_losses = np.sum(np.diff(aircraft_losses[151:243]))

print("Number of aircraft shot down in the summer of 2022:", summer_losses)
```

```python
import pandas as pd

data = pd.read_csv('russia_losses_equipment.csv')

last_300_days_aircraft = data['aircraft'].tail(300)

average_last_300_days = last_300_days_aircraft.mean()

print("Average aviation losses over the last 300 days:", average_last_300_days)
```

```python
import matplotlib.pyplot as plt
import numpy as np

aircraft_losses_first_year = aircraft_losses[:365]

daily_losses_first_year = np.abs(np.diff(aircraft_losses_first_year, prepend=aircraft_losses_first_year[0]))

plt.figure(figsize=(10, 6), dpi=150)

plt.plot(daily_losses_first_year, color='blue', marker='o', linestyle='-', linewidth=2, markersize=4, label='Daily Aircraft Losses')

plt.title('Daily Aircraft Losses During the First Year of the War', fontsize=16, fontweight='bold')
plt.xlabel('Day', fontsize=12)
plt.ylabel('Number of Aircraft Lost', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='upper right')

plt.savefig('daily_aircraft_losses_first_year.png')

plt.show()

print("Values for the graph of daily aircraft losses for the first year of the war: {}.".format(len(daily_losses_first_year)))
print(daily_losses_first_year)
```
